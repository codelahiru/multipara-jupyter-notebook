```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset
path_1 = "https://raw.githubusercontent.com/codelahiru/datasci-datasets/main/Feren/20231124.csv"
df_1 = pd.read_csv(path_1, header=None)

# Set the maximum number of rows and columns to be displayed
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Display the full DataFrame
pd.options.display.width = 1000

df_1.head()
```

```python
# Combine the first three rows into a single row and set it as the header
des_rows_df_1 = df_1.iloc[:3, :]
des_df_1 = des_rows_df_1.apply(lambda x: ' '.join(map(lambda s: '' if pd.isna(s) else str(s), x)), axis=0)

df_1 = df_1[3:]
df_1.columns = des_df_1

# # Convert the data type of header to string
df_1.iloc[:0] = df_1.iloc[:0].astype(str)

# # Display the data types of the header
data_types_df_1_cov = df_1.iloc[:0].dtypes
#print(data_types_df_1_cov)

# Reset the index
df_1.reset_index(drop=True, inplace=True)

# Display the updated DataFrame
df_1.head()
```
